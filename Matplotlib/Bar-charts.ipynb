```python
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25
```

```python
plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
plt.show()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]
```

```python
plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="skyblue")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center')

plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Selected years
years = [2005, 2008, 2011, 2014, 2017]

# Hypothetical runs data
sachin = [1450, 1200, 756, 0, 0]
kohli = [0, 159, 1381, 1054, 1460]
sehwag = [1325, 1220, 933, 0, 0]

# -----------------------------
# Side-by-Side Bar Chart
# -----------------------------
x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(10, 6))

plt.bar(x - width, sachin, width=width, label='Sachin Tendulkar')
plt.bar(x, kohli, width=width, label='Virat Kohli')
plt.bar(x + width, sehwag, width=width, label='Virender Sehwag')

plt.title('Runs Comparison of Sachin, Kohli and Sehwag')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.xticks(x, years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# -----------------------------
# Horizontal Bar Chart
# Total Runs in Debut 5 Years
# -----------------------------

players = ['Sachin', 'Kohli', 'Sehwag']

# Hypothetical total runs scored in first 5 international years
debut_runs = [4200, 3800, 3900]

plt.figure(figsize=(8, 5))

plt.barh(players, debut_runs)

plt.title('Total Runs in First 5 Career Years')
plt.xlabel('Runs')
plt.ylabel('Player')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
